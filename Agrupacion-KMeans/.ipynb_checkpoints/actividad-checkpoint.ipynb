```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
import seaborn as sns
import warnings
warnings.filterwarnings(action="ignore")
plt.style.use('seaborn')
import factor_analyzer as factor
import missingno as msngo
from sklearn.cluster import KMeans
from skimage import io
```

```python
## 2- Importaci贸n de imagenes
tmp_1 = io.imread("album_covers\carly-rae-jepsen-emotion.png")
tmp_2 =  io.imread("album_covers\carly-rae-jepsen-emotion.png")

tmp_1.shape # Son 3 dimenciones que forman un cubro, cada pixel forma 1 color 
```

```python
io.imshow(tmp_1) # Muestra la imagen
```

```python
# Reducci贸n a 16 imagenes

alto, ancho, canal = tmp_2.shape
```

```python
# kmean funciona siempre con 2 dimensiones, se est谩 transformando en 2 dimensiones
datos_procesados = tmp_2.reshape(alto * ancho, canal)
```

```python
# Datos procesados es una matriz con 2 dimensiones
kmeans = KMeans(n_clusters = 16).fit(datos_procesados)
```

```python
# lo que vamos a ver es cuales de estos 16 cluster es el que mas se semeja
datos_procesados[0]
```

```python
# Se generan una matriz de los 16 colores mas representativos
kmeans.cluster_centers_
```

```python
new_colors_pred = kmeans.cluster_centers_[kmeans.predict(datos_procesados)].astype(int)
```

```python
# Cuenta como se representan la cantidad de pixeles en base a 16
pd.DataFrame(new_colors_pred)[0].value_counts()
```

```python
# Convierte en hexadecimal, para que los 3 colores queden en uno solo
def rgb_hex_it(rgb):
    tmp = list(map(lambda x: int(x), rgb))
    return "#%02x%02x%02x" % (tmp[0], tmp[1], tmp[2])
```

```python
hex_convert = list(map(lambda x: rgb_hex_it(x), new_colors_pred))
unique, count = np.unique(kmeans.labels_, return_counts= True)
```

```python
demo = pd.concat([pd.Series(hex_convert), pd.Series(count)], axis = 1)
demo.columns = ["hex", "count"]
```

```python
demo["percentaje"] = demo["count"] / np.sum(demo["count"])
```

```python
hex_percentage = demo["hex"].value_counts("%")
```

```python
hex_percentage
```

```python
plt.barh(hex_percentage.index,
        hex_percentage.values,
        color=hex_percentage.index)
```
