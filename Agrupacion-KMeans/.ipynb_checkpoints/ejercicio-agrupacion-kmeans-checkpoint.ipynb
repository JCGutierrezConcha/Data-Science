## Ejercicio: Agrupación utilizando KMeans y MiniBatchKMeans.

    * Las imágenes se encuentran en una carpeta con el nombre `album_covers`
    * Cada imagen tiene la siguiente nomenclatura: `artista-nombre-del-album.jpg`.
* El objetivo es generar un método que nos permita identificar la dominancia de una cantidad finita de colores.
* Para importar imágenes y visualizarlas, vamos a importar las siguientes librerías:
    * Partamos por incluír las librerías clásicas: `pandas`, `numpy` y `matplotlib.pyplot`.
    * `sklearn.cluster.KMeans` y `sklearn.cluster.MiniBatchKMeans`: para extraer los principales componentes de una matriz numérica.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action="ignore")


from sklearn.cluster import KMeans, MiniBatchKMeans
from skimage import io
```

### 3. Preprocesamiento de imágenes con KMeans y MiniBatchKMeans.

* Con la representación numérica de la imágen, vamos a extraer la altura, el ancho y la cantidad de canales mediante `shape`.
* Posteriormente redimensionaremos la imágen con `reshape`.
* Partamos por inicializar nuestro algoritmo `KMeans` con un `k=8`, ¿Qué significará esto?
* Vuelva a implementar el mismo algoritmo con `MiniBatchKMeans`. ¿Qué diferencia existe con `KMeans`?

```python
img1.shape
```

```python
alto, ancho, canal=img1.shape
```

```python
datos_procesados= img1.reshape(alto*ancho, canal)
```

```python
datos_procesados.shape
```

```python
datos_procesados[0]
```

```python
datos_procesados[1]
```

### Kmeans

```python
kmeans = KMeans(n_clusters=16, random_state=9605).fit(datos_procesados)
```

```python
type(kmeans)
```

```python
pd.Series(kmeans.labels_).value_counts('%')
```

### MiniBatchKMeans

```python
mbkmeans=MiniBatchKMeans(n_clusters=16, random_state=9605).fit(datos_procesados)
```

```python
type(mbkmeans)
```

```python
pd.Series(mbkmeans.labels_).value_counts('%')
```

### 4. Extracción de valores.

* Ahora extraeremos las etiquetas predichas con `labels_`. Hasta el momento las etiquetas hacen referencia a cada centroide. Para imputar sentido en éstos, debemos extraer los valores de los centroides.
* Para extraer los centroides (valores característicos), utilicemos el atributo `cluster_centers_`.
* Con las etiquetas, generaremos un conteo de ocurrencia con `np.unique`. Para extraer el conteo, debemos implementar la opción `return_counts=True`.

```python
labels=kmeans.labels_
```

```python
centers=kmeans.cluster_centers_
```

```python
centers
```

```python
unique, count= np.unique(labels, return_counts=True)
```

```python
unique
```

```python
count
```

```python
porcentaje=count/np.sum(count)
```

```python
porcentaje
```

### 5. Conversión rgb a hex.

* Con los centroides, vamos a convertirlos a formato hexadecimal. 
* Vamos a generar una función y la pasaremos con `map` por cada centroide.

```python
def rgb_hex_it(rgb):
    tmp=list(map(lambda x: int(x), rgb))
    return "#%02x%02x%02x" % (tmp[0], tmp[1], tmp[2])
```

```python
rgb1=[25.2, 12.5,51.3]
rgb_hex_it(rgb1)
```

```python
hex_convert = list(map(lambda x: rgb_hex_it(x), centers))
```

### 6. Definición de base.

* Ahora generaremos un `DataFrame` con las siguientes variables:
    - El color `hex`.
    - La ocurrencia del color en cada pixel `count`.
    - El porcentaje de ocurrencia de cada color respecto a `cluster_centers_`.
* Posteriormente ordenaremos los colores de forma descendente por el porcentaje de ocurrencia.

```python
demo=pd.concat([pd.Series(hex_convert), pd.Series(count), pd.Series(porcentaje)], axis=1)
demo.columns=["hex","count", "porcentaje"]
```

```python
demo=demo.sort_values(by="porcentaje")
demo
```

## MiniBatchKmeans

```python
labels1=mbkmeans.labels_
```

```python
centers1=mbkmeans.cluster_centers_
```

```python
unique1, count1= np.unique(labels1, return_counts=True)
```

```python
porcentaje1=count1/np.sum(count1)
```

```python
hex_convert1 = list(map(lambda x: rgb_hex_it(x), centers1))
```

```python
demo1=pd.concat([pd.Series(hex_convert1), pd.Series(count1), pd.Series(porcentaje1)], axis=1)
demo1.columns=["hex","count","porcentaje"]
```

```python
demo1=demo1.sort_values(by="porcentaje")
demo1
```

### 7. Visualización.

* Genere un gráfico de barras donde presente el porcentaje de cada color. Las barras deben estar coloreadas con el color inferido.


```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Kmeans")
plt.barh(demo["hex"],
         demo["porcentaje"],
         color=demo["hex"])
plt.subplot(1,2,2)
plt.title("MiniBatchKmeans")
plt.barh(demo1["hex"],
         demo1["porcentaje"],
         color=demo1["hex"])
plt.tight_layout()
```
