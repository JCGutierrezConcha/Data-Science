## Ejercicio: Predicción rotación de clientes con Adaboost y GradientBoosting .

Para este ejercicio trabajaremos con una base de datos sobre rotación de clientes en una compañia de telecomunicaciones. 

El archivo contiene 3333 registros y 20 atributos. 

El vector objetivo a modelar es la tasa de rotación entre los clientes de una compañia de telecomunicaciones, que se encuentra en la columna  `churn`.

Los atributos existentes hacen referencia a características de la cuenta de cada cliente.

Lista de atributos:

* State: Estado de Estados Unidos
* Account Length: Tiempo en que la cuenta ha sido activada
* Area Code: código de área.
* international plan: Plan internacional activiado
* voice mail plan: Plan de mensajes de voz activado
* number_vmail_messages: Cantidad de mensajes de voz.
* total_day_minutes: Cantidad de minutos ocupados en la mañana.
* total_day_calls: Cantidad de llamadas realizadas en la mañana.
* total_day_charge: Cobros realizados en la mañana.
* total_eve_minutes: Cantidad de minutos ocupados en la tarde.
* total_eve_calls: Cantidad de llamadas realizadas en la tarde.
* total_eve_charge: Cobros realizados en la tarde.
* total_night_calls: Cantidad de llamadas realizadas en la noche.
* tota_night_minutes: Cantidad de minutos ocupados en la noche.
* total_night_charge: Cobros realizados en la noche.
* total_intl_minutes: Cantidad de minutos ocupados en llamadas internacionales.
* total_intl_calls: Cantidad de llamadas internacionales realizadas.
* total_intl_charge: Cobros realizados por llamadas internacionales.
* churn: 1 si el cliente se cambió de compañia, 0 de los contrario.

Los datos provienen del paquete AppliedPreditiveModeling de R .

### 2. Comparación entre AdaBoost y Gradient Boosting.

Entrene los modelos de clasificación AdaBoost y Gradient Boosting para mejorar su capacidad predictiva en la medida de lo posible.

Para ello, implemente una búsqueda de grilla con las siguientes especificaciones:

AdaBoostClassifier: 
- learning_rate : [0.01, 0.1, 0.5]
- n_estimators  : [50, 100, 500, 1000, 2000]

GradientBoostingClassifier: 
- learning_rate: [0.01, 0.1,0.5]
- n_estimators : [50, 100, 500, 1000,2000]
- subsample    : [0.1,0.5,0.9]


Si el tiempo de computación es alto, puede implementar la búsqueda con 3 validaciones cruzadas.

Reporte las métricas para los mejores modelos.

```python
from sklearn.model_selection import  train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
```

```python
y= df['churn_yes']
X= df.drop(columns='churn_yes')
n = .33
seed = 9605
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=n, random_state=seed)
```

```python
params_adaboost={'learning_rate': [0.01, 0.1, 0.5], 'n_estimators': [50, 100, 500, 1000, 2000]}
params_gradientboost={'learning_rate': [0.01, 0.1,0.5],'n_estimators': [50, 100, 500, 1000,2000],'subsample':[0.1,0.5,0.9]}
```

### AdaBoost

```python
%%time
adaboost_grid_cv = GridSearchCV(AdaBoostClassifier(random_state=seed),
                                param_grid=params_adaboost,
                                cv=10,
                                n_jobs=-1).fit(X_train, y_train)
```

```python
# Mejor combinación de hiperparámtros.
print("Mejor Combinacion de Hiperparámetros AdaBoost: {}".format(adaboost_grid_cv.best_params_))
```

```python
# Mejor modelo.
adaboost_best=adaboost_grid_cv.best_estimator_.fit(X_train, y_train)
```

```python
adaboost_best
```

```python
# Métricas de desempeño.
y_pred= adaboost_best.predict(X_test)
crep_adaboost=pd.DataFrame(classification_report(y_test, y_pred, target_names=['No Fuga','Fuga'],output_dict=True)).drop(columns=['accuracy', 'macro avg']).head(4)
fn.model_metrics(crep_adaboost)
```

### Comentario sobre desempeño del Clasificador AdaBoost.

Este modelo (Clasificador AdaBoost), presenta un desempeño insatisfactorio para clasificar la clase 'Fuga', con valores bajos de f1-score (0.347) y Recall (0.258).

Por otra parte, el modelo presenta un buen desempeño para clasificar la clase 'No Fuga', con valores sobre 0.884 en las 3 métricas consideradas (f1-score, Precision y Recall).

### Gradient Boosting

```python
%%time
gradientboost_grid_cv = GridSearchCV(GradientBoostingClassifier(random_state=seed),
                                param_grid=params_gradientboost,
                                cv=10,
                                n_jobs=-1).fit(X_train, y_train)
```

```python
# Mejor combinación de hiperparámtros.
print("Mejor Combinacion de Hiperparámetros GradientBoosting: {}".format(gradientboost_grid_cv.best_params_))
```

```python
# Mejor modelo.
gradientboost_best=gradientboost_grid_cv.best_estimator_.fit(X_train, y_train)
```

```python
gradientboost_best
```

```python
# Métricas de desempeño.
y_pred= gradientboost_best.predict(X_test)
crep_gradientboost=pd.DataFrame(classification_report(y_test, y_pred, target_names=['No Fuga','Fuga'],output_dict=True)).drop(columns=['accuracy', 'macro avg']).head(4)
fn.model_metrics(crep_gradientboost)
```

### Comentario sobre desempeño del Clasificador Gradient Boosting.

Este modelo (Clasificador Gradient Boosting), presenta un buen desempeño para clasificar la clase 'Fuga', con valores sobre 0.735 en las 3 métricas consideradas (f1-score, Precision y Recall).

Por otra parte, el modelo también presenta un buen desempeño para clasificar la clase 'No Fuga', con valores sobre 0.956 en las 3 métricas consideradas (f1-score, Precision y Recall).

Se observa claramente, que el desempeño de Gradient Boosting es superior al que presenta el modelo AdaBoost, en particular en lo que respecta a las métricas para clasificar la clase 'Fuga', en que f1-score aumenta de 0.347 a 0.815, Recall aumenta de 0.258 a 0.736 y Precision aumenta de 0.532 a 0.914.

```python
def order_features(best_model, X_features, k_feats=10):
    ord_feat=sorted(zip(map(lambda x: round(x, 4), best_model.feature_importances_), X_features), reverse=True)[:k_feats]
    suma=0
    for i in ord_feat:
        suma +=i[0]
    print("Porcentaje que cubren los {} principales atributos es: {}%". format(k_feats,round(suma*100,1)))
    print("Features sorted by their score:")
    return ord_feat
```
