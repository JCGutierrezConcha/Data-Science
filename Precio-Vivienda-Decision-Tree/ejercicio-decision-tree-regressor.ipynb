```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import warnings
warnings.filterwarnings('ignore')

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import median_absolute_error, mean_squared_error
from sklearn.model_selection import  train_test_split
from sklearn.model_selection import GridSearchCV
```

### 2. Feature engineering.

Identifique si el dtype de cada pd.Serie en nuestra base de datos se considera 'object' o no. 

Para todas las variables que sean 'object' , realice lo siguiente:

* Genere una recodificación $K-1$ en cada variable. Para efectos prácticos sólo necesitan eliminar una de las categorías, no se concentren en especificar la categoría a eliminar. Pueden utilizar la función pd.get_dummies con la opción drop_first para ello.

* Utilizando el método pd.concat , concatene a los atributos creados en la base de datos.

tip: No se olvide de eliminar los atributos recodificados, de esta forma evitará un aumento artificial del desempeño del modelo.

```python
# Identificación dtype de cada Serie.
df.info()
```

```python
# Selección de atributos con dtyte=object
col_object=[]
for colnames, serie in df.iteritems():
    if serie.dtype == 'object':
        col_object.append(colnames)
```

```python
len(col_object)
```

```python
# Recodificación K-1
for col in col_object:
    df = pd.concat([df, pd.get_dummies(df[col], drop_first=True, prefix= col)], axis=1)
    df = df.drop(columns=col)
```

```python
len(df.columns)
```

### 3. Primer modelo con Decision Tree Regressor.

Genere muestras de entrenamiento y validación con 'Sale_Price' como vector objetivo y los atributos de la base de datos como matriz.

Recuerde definir el porcentaje de casos en la muestra de validación y una semilla pseudoaleatoria.

```python
y= df['Sale_Price']
X= df.drop(columns='Sale_Price')
n = .33
seed = 9605
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=n, random_state=seed)
```

### 4. Importancia relativa.

Implemente el método plot_importance para reportar la importancia relativa de los atributos.

Comente sobre cuáles son los principales 10 atributos que afectan la predicción de Sale_Price .

Separe éstos 10 atributos en una nueva base de datos, junto con el vector objetivo.

```python
def plot_feature_importance(fit_model, feat_names):
    """
    Plot relative importance of a feature subset given a fitted model.
    """
    # infer feature importance score
    tmp_importance = fit_model.feature_importances_
    # sort features
    sort_importance = np.argsort(tmp_importance)[::-1]
    # associate feat_names with its relative importance
    names = [feat_names[i] for i in sort_importance]
    # plot
    plt.barh(
        # given range and features
        range(len(feat_names)), tmp_importance[sort_importance]
    )
    # add axis labels identifying attribute name
    plt.yticks(range(len(feat_names)), 
               names, rotation=0)
```

```python
plt.figure(figsize=(15,70))
plot_feature_importance(reg_tree, X.columns)
```

```python
def infer_k_features(df, model, feat_names, k_feats=10):
    """TODO: Docstring for infer_k_features.
    :arg1: TODO
    :returns: TODO
    """
    # preserve temporary copy
    tmp_df = df.copy()
    # infer feature importance score
    tmp_importance = model.feature_importances_
    # sort features
    sort_importance = np.argsort(tmp_importance)[::-1]
    # associate feat names with its relative importance
    names = [feat_names[i] for i in sort_importance]
    #mungle into a dataframe
    tmp_attr = pd.DataFrame(
        {'name': names,
         'score': tmp_importance[sort_importance]}
    )
    # restrict dataframe to k attributes
    tmp_attr = tmp_attr[:k_feats]['name']
    
    list_attr=[]
    for n in tmp_attr:
        list_attr.append(n)

    # filter attributes
    #tmp_df = tmp_df[tmp_df.columns[tmp_df.columns.isin(tmp_attr)]]
    tmp_df = tmp_df.loc[:,list_attr]
    

    return tmp_df
```

```python
df10=infer_k_features(df, reg_tree, X.columns, k_feats=10)
```

```python
def order_features(best_model, X_features, k_feats=10):
    ord_feat=sorted(zip(map(lambda x: round(x, 4), best_model.feature_importances_), X_features), reverse=True)[:k_feats]
    suma=0
    for i in ord_feat:
        suma +=i[0]
    print("Porcentaje que cubren los {} principales atributos es: {}%". format(k_feats,round(suma*100,1)))
    print("Features sorted by their score:")
    return ord_feat
```

### 5. Refactorización del modelo y pickling.

En función de los atributos seleccionados en el ejercicio anterior, vuelva a generar conjuntos de entrenamiento y validación.

Dentro de los datos de entrenamiento genere una búsqueda de grilla con GridSearchCV utilizando los siguientes hiperparámetros:

* Máximo de atributos: Evalúe todos los posibles atributos.
* Máximo de profundidad: entre 1 a 32.
* Validaciones cruzadas : 5.

Reporte la mejor combinación de hiperparámetros y su desempeño asociado. Compare el desempeño en la muestra de validación con el modelo por defecto.


```python
y1= df10['Sale_Price']
X1= df10.drop(columns='Sale_Price')
n = .33
seed = 9605
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=n, random_state=seed)
```

```python
%%time
reg_tree_grid_cv = GridSearchCV(DecisionTreeRegressor(random_state=seed),
                                {  # con una profundidad de ramas hasta 32
                                   'max_depth': np.linspace(1, 32, 32),
                                   # evaluando todos los atributos en la matriz
                                   'max_features': list(range(1,X1.shape[1]+1))},
                                # Con 5 validaciones cruzadas
                                cv=5,
                                # Ocupando todos los núcleos del computador
                                n_jobs=-1).fit(X1_train, y1_train)
```

```python
# Mejor combinación de hiperparámtros.
print("Mejor Combinacion de Hiperparámetros Arbol de Regresión: {}".format(reg_tree_grid_cv.best_params_))
```

### Desempeño del modelo con hiperparámetros.

```python
reg_tree_best=reg_tree_grid_cv.best_estimator_
```

```python
reg_tree_best
```

```python
reg_tree_best=reg_tree_best.fit(X1_train, y1_train)
```

```python
mae_reg_tree_best =median_absolute_error( y1_test, reg_tree_best.predict(X1_test)).round(1)
print('MAE Regresion Tree Best: {0}'.format(mae_reg_tree_best))
```

```python
mse_reg_tree_best=mean_squared_error( y1_test, reg_tree_best.predict(X1_test)).round(1)
print('MSE Regresion Tree Best: {0}'.format(mse_reg_tree_best))
```

### Comparación desempeño entre modelos

```python
plt.figure(figsize=(8,3))
plt.barh(['Modelo 1\nRegresion Tree Base', 'Modelo 2\nRegresion Tree con 10 principales atributos'], [mse_reg_tree, mse_reg_tree_best])
plt.title("Mean Square Error (MSE)")
plt.tight_layout();
```

```python
varia_mse= ((mse_reg_tree_best-mse_reg_tree)/mse_reg_tree*100).round(1)
print('Variación MSE (%): {0}'.format(varia_mse))
```

```python
plt.figure(figsize=(8,3))
plt.barh(['Modelo 1\nRegresion Tree Base', 'Modelo 2\nRegresion Tree con 10 principales atributos'], [mae_reg_tree, mae_reg_tree_best])
plt.title("Median Absolute Error (MAE)")
plt.tight_layout();
```

```python
varia_mae= ((mae_reg_tree_best-mae_reg_tree)/mae_reg_tree*100).round(1)
print('Variación MAE (%): {0}'.format(varia_mae))
```
